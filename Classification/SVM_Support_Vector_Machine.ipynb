### &#128279; <span style="color:#4682b4;">Support Vector Machine (SVM)</span>

**Support Vector Machine (SVM)** is a powerful supervised learning algorithm used for classification and regression tasks. It is particularly effective in high-dimensional spaces and for problems where the number of dimensions exceeds the number of samples.

#### **How SVM Works**:
1. **Hyperplane**: SVM constructs a hyperplane or a set of hyperplanes in a high-dimensional space to separate different classes. The goal is to find the hyperplane that best separates the classes with the maximum margin.
2. **Support Vectors**: The data points that are closest to the hyperplane are called support vectors. These points are critical in defining the position and orientation of the hyperplane.
3. **Maximize Margin**: The margin is the distance between the hyperplane and the nearest support vectors. SVM aims to maximize this margin to achieve better generalization on unseen data.

#### **Mathematical Representation**:
- **Hyperplane**: $ w \cdot x + b = 0 $, where $w$ is the weight vector and $b$ is the bias.
- **Decision Function**: For a given data point $ x $, the decision function is $ f(x) = \text{sign}(w \cdot x + b) $.
- **Objective**: Minimize $ \frac{1}{2} ||w||^2 $ subject to $ y_i (w \cdot x_i + b) \geq 1 $ for all $ i $, where $ y_i $ is the class label of the $ i^{th} $ sample.

#### **Advantages of SVM**:
- **Effective in high-dimensional spaces**: SVM works well with a large number of features.
- **Robust to overfitting**: Especially in high-dimensional space, SVM is less prone to overfitting.

#### **Disadvantages of SVM**:
- **Computationally intensive**: Training time can be long, especially with large datasets.
- **Choice of kernel**: The performance of SVM depends on the choice of the kernel and its parameters.

#### **Example of SVM Classification**:
Suppose you have a dataset of email messages labeled as spam or not spam. SVM will find the hyperplane that maximally separates the spam emails from the non-spam emails. The support vectors in this case are the emails that are closest to the boundary between the two classes.

For more information, please visit the [SVM Wikipedia page](https://en.wikipedia.org/wiki/Support_vector_machine).

### ðŸŒ¸ Iris Dataset

**The Iris dataset** is a classic dataset in the field of machine learning and statistics, often used for pattern recognition, classification, and clustering tasks. It contains 150 samples of iris flowers, each described by four features:

- **Sepal Length**: The length of the sepal in centimeters.
- **Sepal Width**: The width of the sepal in centimeters.
- **Petal Length**: The length of the petal in centimeters.
- **Petal Width**: The width of the petal in centimeters.

The dataset is divided into three classes, each representing a different species of iris flower:

- **Iris-setosa**
- **Iris-versicolor**
- **Iris-virginica**

Each class has 50 samples, making it a balanced dataset. The simplicity and small size of the Iris dataset make it an ideal candidate for testing and comparing different machine learning algorithms.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Convert to a DataFrame for easier manipulation
iris_df = pd.DataFrame(data=X, columns=feature_names)
iris_df['target'] = y
iris_df['target_name'] = iris_df['target'].apply(lambda x: target_names[x])
```

```python
# Set up the matplotlib figure
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Scatter Plots of Iris Dataset Features', fontsize=16)

# Create a list of feature combinations
feature_combinations = [(i, j) for i in range(len(feature_names)) for j in range(i + 1, len(feature_names))]

# Plot scatter plots
for idx, (i, j) in enumerate(feature_combinations):
    row, col = divmod(idx, 3)
    ax = axes[row, col]
    sns.scatterplot(x=iris_df[feature_names[i]], y=iris_df[feature_names[j]], hue=iris_df['target_name'], palette='deep', ax=ax)
    ax.set_xlabel(feature_names[i] , size = 16)
    ax.set_ylabel(feature_names[j] , size = 16)
    ax.legend(loc='upper right')

# Hide any unused subplots
if len(feature_combinations) < 6:
    for idx in range(len(feature_combinations), 6):
        row, col = divmod(idx, 3)
        fig.delaxes(axes[row, col])

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

### &#9881; <span style="color:#4682b4;">Hyperparameters of Support Vector Machine (SVM)</span>

Support Vector Machines (SVMs) are powerful and versatile supervised learning models used for classification and regression tasks. Tuning the hyperparameters of an SVM is crucial for optimizing its performance. Below are the key hyperparameters of SVM:

#### **1. C (Regularization Parameter)**
- **Description**: The regularization parameter $C$ controls the trade-off between achieving a low training error and a low testing error (generalization).
- **Effect**: A small value of $C$ makes the decision surface smooth, while a large value of $C$ aims to classify all training examples correctly.
- **Formula**: The objective function of SVM includes a regularization term:
  $$
  \text{Minimize} \quad \frac{1}{2} ||w||^2 + C \sum_{i=1}^{n} \xi_i
  $$
  where $w$ is the weight vector, $\xi_i$ are the slack variables, and $C$ is the regularization parameter.
- **Default Value**: 1.0

#### **2. Kernel**
- **Description**: The kernel parameter specifies the kernel type to be used in the algorithm. SVM supports several kernels:
  - `linear`: Linear kernel
  - `poly`: Polynomial kernel
  - `rbf`: Radial Basis Function (Gaussian) kernel
  - `sigmoid`: Sigmoid kernel
  - `precomputed`: A kernel matrix is precomputed
- **Effect**: Different kernels can capture different patterns in the data.
- **Default Value**: `rbf`

#### **3. Degree**
- **Description**: The degree of the polynomial kernel function (`poly`). Ignored by other kernels.
- **Effect**: A higher degree can capture more complex patterns but may lead to overfitting.
- **Default Value**: 3

#### **4. Gamma**
- **Description**: The kernel coefficient for `rbf`, `poly`, and `sigmoid`.
- **Effect**: A small value of gamma makes the model behave like a linear model, while a large value of gamma leads to overfitting.
- **Options**: `scale` (default), `auto`, or a float value.
  - `scale`: $\frac{1}{\text{number of features} \times \text{variance of features}}$
  - `auto`: $\frac{1}{\text{number of features}}$
- **Default Value**: `scale`

#### **5. Coef0**
- **Description**: The independent term in kernel function. It's used in `poly` and `sigmoid` kernels.
- **Effect**: Adjusts the influence of higher-order versus lower-order terms in the polynomial and sigmoid kernels.
- **Default Value**: 0.0

#### **6. Probability**
- **Description**: Whether to enable probability estimates. This must be enabled prior to calling `fit`, and will slow down that method.
- **Effect**: It allows the model to predict probabilities of each class.
- **Default Value**: False

#### **7. Shrinking**
- **Description**: Whether to use the shrinking heuristic.
- **Effect**: It can speed up the optimization process.
- **Default Value**: True

#### **8. Class Weight**
- **Description**: Set the parameter $C$ of class $i$ to `class_weight[i] * C` for SVC. If not given, all classes are supposed to have weight one.
- **Effect**: Helps to handle imbalanced datasets by assigning different weights to classes.
- **Options**: 'balanced' or a dictionary of class weights.
- **Default Value**: None

#### **9. Verbose**
- **Description**: Enable verbose output.
- **Effect**: Provides detailed output for debugging.
- **Default Value**: False

#### **10. Max Iter**
- **Description**: Hard limit on iterations within solver, or -1 for no limit.
- **Effect**: Controls the number of iterations the algorithm will run.
- **Default Value**: -1

#### **11. Random State**
- **Description**: Controls the pseudo-random number generation for shuffling the data for probability estimates.
- **Effect**: Ensures reproducibility of the results.
- **Default Value**: None

By carefully tuning these hyperparameters, you can optimize the performance of your SVM model for your specific problem.

</div>

```python
# Import necessary libraries
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Prepare feature and target data
X = iris_df.iloc[:, :2].values
y = iris_df['target'].values

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize lists to store accuracy for different C values
C_values = np.linspace(0.01 , 1 , 10)
accuracy_scores = []

# Train and evaluate SVM classifier for different C values
for C in C_values:
    svm = SVC(C=C ,gamma = 'scale')
    svm.fit(X_train, y_train )
    y_pred = svm.predict(X_test)
    accuracy_scores.append(accuracy_score(y_test, y_pred))

# Print accuracy scores for different C values
for C, score in zip(C_values, accuracy_scores):
    print(f"C={round(C,2)}, Accuracy={score:.4f}")
```

```python
# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(C_values, accuracy_scores, marker='o', color='darkblue')
plt.xlabel('Regularization Parameter (C)')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Regularization Parameter (C) for SVM')
plt.xticks(C_values)
plt.grid(True)
plt.show()
```

```python
# Fit KNN model with k=1
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

# Create a mesh grid for decision boundary plotting
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Plot the decision boundaries for both models
plt.figure(figsize=(10, 5))

# Plot for SVM (C=1)
plt.subplot(1, 2, 1)
Z_svm = svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z_svm = Z_svm.reshape(xx.shape)
plt.contourf(xx, yy, Z_svm, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
plt.title('SVM (C=1) Decision Boundary')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

# Plot for KNN (k=1)
plt.subplot(1, 2, 2)
Z_knn = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z_knn = Z_knn.reshape(xx.shape)
plt.contourf(xx, yy, Z_knn, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
plt.title('KNN (k=1) Decision Boundary')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.tight_layout()
plt.show()
```

### &#128279; <span style="color:#4682b4;">Comparison of SVM (C=1) and KNN (k=1)</span>

After implementing SVM with C=1 and KNN with k=1 on the dataset, we can observe that SVM exhibits a lower complexity and smoother decision boundary compared to KNN. Hereâ€™s an explanation of why this might be the case:

#### **SVM (C=1)**:
- SVM aims to find a decision boundary that maximizes the margin between classes, leading to a more generalized solution.
- With a smaller C value (regularization parameter), SVM tends to prioritize a wider margin, resulting in a smoother decision boundary that is less influenced by individual data points.
- This regularization helps SVM generalize better to unseen data and reduces the risk of overfitting.

#### **KNN (k=1)**:
- KNN, as a non-parametric model, directly learns from the training data and classifies new points based on their proximity to the nearest neighbors.
- When k=1, KNN tends to have a decision boundary that closely follows the training data points, leading to a more intricate boundary.
- The model may capture noise or outliers in the training data, which can result in a decision boundary that is more jagged and complex, potentially leading to overfitting.

#### **Conclusion**:
In our experiment, SVM with C=1 demonstrates a smoother decision boundary and lower complexity compared to KNN with k=1. This characteristic of SVM can be advantageous in scenarios where we prioritize generalization and robustness to unseen data. However, the choice between SVM and KNN ultimately depends on the specific dataset and the balance between bias and variance that best suits the problem at hand.
