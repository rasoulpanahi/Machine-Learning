### &#128279; <span style="color:#4682b4;">Naive Bayes Classifiers</span>

**Naive Bayes** is a family of simple yet effective probabilistic classifiers based on applying Bayes' theorem with the "naive" assumption of conditional independence between every pair of features given the class label. It is commonly used for classification tasks, especially with large datasets.

#### **How Naive Bayes Works**:
1. **Bayes' Theorem**: Naive Bayes classifiers apply Bayes' theorem to compute the probability of a class given a feature vector. The theorem is formulated as:
   $$
   P(C|X) = \frac{P(X|C) \cdot P(C)}{P(X)}
   $$
   where $P(C|X)$ is the posterior probability of class $C$ given the feature vector $X$, $P(X|C)$ is the likelihood of $X$ given class $C$, $P(C)$ is the prior probability of class $C$, and $P(X)$ is the probability of the feature vector $X$.

2. **Conditional Independence Assumption**: Naive Bayes assumes that the features are conditionally independent given the class label. Thus, the likelihood $P(X|C)$ can be decomposed as:
   $$
   P(X|C) = \prod_{i=1}^{n} P(x_i|C)
   $$
   where $x_i$ are the individual features.

3. **Classification**: The class with the highest posterior probability is chosen as the predicted class:
   $$
   \hat{C} = \arg\max_{C} P(C|X)
   $$

#### **Mathematical Representation**:
- **Posterior Probability**: $ P(C|X) = \frac{P(X|C) \cdot P(C)}{P(X)} $
- **Likelihood Decomposition**: $ P(X|C) = \prod_{i=1}^{n} P(x_i|C) $
- **Prediction**: $ \hat{C} = \arg\max_{C} P(C|X) $

#### **Advantages of Naive Bayes**:
- **Simple and Fast**: Naive Bayes is computationally efficient and works well with large datasets.
- **Works with Small Data**: Requires less training data to estimate the parameters.
- **Handles Irrelevant Features**: Less sensitive to irrelevant features compared to other classifiers.
- **Scalable**: Can be scaled to large datasets easily.

#### **Disadvantages of Naive Bayes**:
- **Independence Assumption**: The assumption that features are conditionally independent given the class label is often unrealistic, which can affect the model's performance.
- **Zero Probability Problem**: If a feature value was not observed in the training data for a given class, the probability estimate becomes zero, which can be problematic. This can be mitigated with techniques like Laplace smoothing.

#### **Example of Naive Bayes Classification**:
Suppose you have a dataset of emails labeled as spam or not spam. Each email is represented by features such as the presence of specific words. Naive Bayes will compute the probability of an email being spam based on the presence or absence of these words and classify it accordingly.

### &#128279; <span style="color:#4682b4;">Calculating $P(C)$ and $P(X|C)$ in Naive Bayes</span>

**Naive Bayes** classifiers rely on calculating the prior probability of each class ($P(C)$) and the likelihood of the features given the class ($P(X|C)$). Here's how these probabilities are computed:

#### **Calculating $P(C)$ (Prior Probability)**:
The prior probability of class $C$ is the proportion of instances in the dataset that belong to class $C$. It is calculated as:
$$
P(C) = \frac{\text{Number of instances of class } C}{\text{Total number of instances}}
$$

For example, if you have a dataset of emails where 40 out of 100 emails are spam, the prior probability of an email being spam is:
$$
P(\text{spam}) = \frac{40}{100} = 0.4
$$

#### **Calculating $P(X|C)$ (Likelihood)**:
The likelihood $P(X|C)$ is the probability of observing the feature vector $X$ given the class $C$. Since Naive Bayes assumes conditional independence of features, the likelihood can be decomposed into the product of the individual feature probabilities:
$$
P(X|C) = P(x_1, x_2, \ldots, x_n|C) = \prod_{i=1}^{n} P(x_i|C)
$$

To compute $P(x_i|C)$, the method varies depending on the type of data:

1. **For Categorical Data**:
   - Count the number of instances where feature $x_i$ takes a specific value for class $C$.
   - Divide by the total number of instances of class $C$.
   $$
   P(x_i = v|C) = \frac{\text{Number of instances where } x_i = v \text{ and class } C}{\text{Number of instances of class } C}
   $$

2. **For Continuous Data**:
   - Assume a Gaussian (normal) distribution for the feature values.
   - Calculate the mean ($\mu_C$) and variance ($\sigma_C^2$) of the feature $x_i$ for class $C$.
   - Use the probability density function of the normal distribution:
   $$
   P(x_i|C) = \frac{1}{\sqrt{2 \pi \sigma_C^2}} \exp\left(-\frac{(x_i - \mu_C)^2}{2 \sigma_C^2}\right)
   $$

#### **Example Calculation**:
Let's illustrate with a simple example using categorical data. Suppose you have a dataset of fruits labeled as "apple" or "orange" with features "color" and "size".

- **Prior Probability**:
  - $P(\text{apple}) = \frac{\text{Number of apples}}{\text{Total number of fruits}}$
  - $P(\text{orange}) = \frac{\text{Number of oranges}}{\text{Total number of fruits}}$

- **Likelihood**:
  - For feature "color", $P(\text{color}=\text{red}|\text{apple}) = \frac{\text{Number of red apples}}{\text{Total number of apples}}$
  - For feature "size", $P(\text{size}=\text{small}|\text{orange}) = \frac{\text{Number of small oranges}}{\text{Total number of oranges}}$

By calculating these probabilities, Naive Bayes can estimate the posterior probability of a fruit being an apple or an orange given its color and size.

For more information, please visit the [Naive Bayes Wikipedia page](https://en.wikipedia.org/wiki/Naive_Bayes_classifier).

### 🌸 Iris Dataset

**The Iris dataset** is a classic dataset in the field of machine learning and statistics, often used for pattern recognition, classification, and clustering tasks. It contains 150 samples of iris flowers, each described by four features:

- **Sepal Length**: The length of the sepal in centimeters.
- **Sepal Width**: The width of the sepal in centimeters.
- **Petal Length**: The length of the petal in centimeters.
- **Petal Width**: The width of the petal in centimeters.

The dataset is divided into three classes, each representing a different species of iris flower:

- **Iris-setosa**
- **Iris-versicolor**
- **Iris-virginica**

Each class has 50 samples, making it a balanced dataset. The simplicity and small size of the Iris dataset make it an ideal candidate for testing and comparing different machine learning algorithms.

### &#9881; <span style="color:#4682b4;">Hyperparameters of Naive Bayes Classifier</span>

Naive Bayes classifiers are a family of simple and effective probabilistic classifiers based on applying Bayes' theorem with strong (naive) independence assumptions between the features. Tuning the hyperparameters of a Naive Bayes classifier can significantly affect its performance. Below are the key hyperparameters of the most common Naive Bayes classifiers:

#### **1. Alpha (Laplace Smoothing)**
- **Description**: The smoothing parameter, used primarily in Multinomial and Bernoulli Naive Bayes, to handle zero probabilities.
- **Effect**: Adding a small positive number $\alpha$ to the frequency counts prevents zero probabilities in the presence of previously unseen feature values.
- **Formula**: 
  $$
  P(x_i|C) = \frac{\text{count}(x_i, C) + \alpha}{\text{count}(C) + \alpha \cdot n}
  $$
  where $n$ is the number of possible feature values.
- **Default Value**: 1.0

#### **2. Fit Prior**
- **Description**: Whether to learn class prior probabilities or not. If false, a uniform prior is used.
- **Effect**: Determines whether the classifier should consider the relative frequency of classes in the training data or assume equal class probabilities.
- **Default Value**: True

#### **3. Class Prior**
- **Description**: Prior probabilities of the classes. If specified, the priors are not adjusted according to the data.
- **Effect**: Allows the user to set their own prior probabilities for the classes, which can be useful if domain knowledge about class distribution is available.
- **Default Value**: None

#### **4. Binarize**
- **Description**: Threshold for binarizing (mapping to 0/1) of the input features. Used in Bernoulli Naive Bayes.
- **Effect**: Converts continuous features into binary features. If set to a float, all values above the threshold are mapped to 1, and all values below are mapped to 0.
- **Default Value**: 0.0

#### **5. Priors**
- **Description**: Prior probabilities of the classes. This is used in Gaussian Naive Bayes to set the prior probabilities.
- **Effect**: Adjusts the decision boundary based on the prior knowledge of class distribution.
- **Default Value**: None

#### **6. var_smoothing (Variance Smoothing)**
- **Description**: Portion of the largest variance of all features that is added to variances for calculation stability in Gaussian Naive Bayes.
- **Effect**: Helps to avoid division by zero or other numerical instabilities by adding a small value to the variance.
- **Default Value**: 1e-9

#### **Advantages of Naive Bayes**:
- **Fast and Efficient**: Naive Bayes classifiers are highly scalable, requiring a number of parameters linear in the number of features.
- **Performs well on small datasets**: They are effective even with a small amount of training data.
- **Handles missing data**: Can be implemented to handle missing data easily.

#### **Disadvantages of Naive Bayes**:
- **Independence Assumption**: The assumption that features are independent is often unrealistic in real-world applications.
- **Zero Probability Problem**: If a category in a feature was not observed in the training data, the model assigns zero probability, though this is addressed by smoothing techniques like Laplace smoothing.

By carefully tuning these hyperparameters, you can optimize the performance of your Naive Bayes model for your specific problem.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Load the iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Prepare feature and target data
X = iris_df.iloc[:, :2].values
y = iris_df['target'].values

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize lists to store accuracy for different alpha values
alpha_values = np.linspace(0.01, 1.1, 10)
accuracy_scores = []

# Train and evaluate Naive Bayes classifier for different alpha values
for alpha in alpha_values:
    nb = MultinomialNB(alpha=alpha)
    nb.fit(X_train, y_train)
    y_pred = nb.predict(X_test)
    accuracy_scores.append(accuracy_score(y_test, y_pred))

# Print accuracy scores for different alpha values
for alpha, score in zip(alpha_values, accuracy_scores):
    print(f"alpha={round(alpha, 2)}, Accuracy={score:.4f}")

# Plot the accuracy scores
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(alpha_values, accuracy_scores, marker='o', color='darkblue')
plt.xlabel('Alpha (Laplace Smoothing)')
plt.ylabel('Accuracy')
plt.title('Effect of Alpha on Accuracy for Naive Bayes Classifier')
plt.xticks(alpha_values)
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix

# Load the iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Prepare feature and target data
X = iris_df.iloc[:, :2].values  # Use only the first two features for 2D visualization
y = iris_df['target'].values

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the Naive Bayes classifier
nb = GaussianNB()
nb.fit(X_train, y_train)

# Predict the class labels for the test set
y_pred = nb.predict(X_test)

# Print the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Compute the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Create a mesh grid
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict class labels for each point in the mesh grid
Z = nb.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and scatter plot
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, edgecolors='k', cmap=plt.cm.coolwarm)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Naive Bayes Decision Boundary (Train Data)')
plt.show()
```
