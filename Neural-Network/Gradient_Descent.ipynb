Gradient Descent is an optimization algorithm used to minimize the cost function in machine learning models. Here's why it is crucial:

#### &#127808; What is Gradient Descent?
Gradient Descent iteratively adjusts the model parameters to reduce the difference between predicted and actual values. It moves in the direction of the steepest decrease of the cost function, hence "descending the gradient".

#### &#128640; Why Do We Need It?
- **Model Optimization**: Gradient Descent helps in finding the optimal parameters for our models, ensuring they make accurate predictions.
- **Efficiency**: By taking small, calculated steps towards the minimum cost, it efficiently navigates the parameter space.
- **Scalability**: It works well with large datasets and complex models, making it a versatile tool in machine learning.

### &#128211; <span style="color:#4682b4;">Mathematical Insights into Gradient Descent</span>

Gradient Descent is a fundamental optimization algorithm used extensively in machine learning. It works by iteratively moving towards the minimum of a cost function, thus refining the model parameters. Letâ€™s delve deeper into the mathematics behind this powerful algorithm.

#### &#128196; The Cost Function
The cost function (or loss function) quantifies the error between the predicted values and the actual values. For a regression problem, the Mean Squared Error (MSE) is commonly used:

$$
J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} \left( h_{\theta}(x^{(i)}) - y^{(i)} \right)^2
$$

Here:
- $m$ is the number of training examples.
- $h_{\theta}(x^{(i)})$ is the hypothesis function, which in linear regression is $h_{\theta}(x) = \theta^T x$.
- $y^{(i)}$ is the actual output.

#### &#128200; Gradient Computation
The gradient of the cost function with respect to each parameter $\theta_j$ indicates the direction and rate of fastest increase of the function. For our MSE cost function, the gradient is:

$$
\frac{\partial J(\theta)}{\partial \theta_j} = \frac{1}{m} \sum_{i=1}^{m} \left( h_{\theta}(x^{(i)}) - y^{(i)} \right) x_j^{(i)}
$$

This gradient is a vector of partial derivatives and points in the direction of the steepest ascent.

#### &#128275; Gradient Descent Update Rule
To minimize the cost function, we move in the direction opposite to the gradient. The update rule for each parameter $\theta_j$ is:

$$
\theta_j := \theta_j - \alpha \frac{\partial J(\theta)}{\partial \theta_j}
$$

Where:
- $\alpha$ is the learning rate, a positive scalar determining the size of the steps taken towards the minimum.

#### &#128176; Learning Rate
Choosing an appropriate learning rate $\alpha$ is crucial:
- If $\alpha$ is too small, the algorithm will take tiny steps and converge slowly.
- If $\alpha$ is too large, the algorithm might overshoot the minimum or even diverge.

#### &#127919; Convergence
Gradient Descent converges to a local minimum of the cost function. In the case of convex functions, like the MSE for linear regression, this local minimum is also the global minimum.

#### &#127882; Variants of Gradient Descent
1. **Batch Gradient Descent**:
   - Uses the entire dataset to compute the gradient.
   - Update rule:

   $$
   \theta := \theta - \alpha \nabla J(\theta)
   $$

2. **Stochastic Gradient Descent (SGD)**:
   - Uses a single training example to compute the gradient.
   - Update rule:

   $$
   \theta := \theta - \alpha \nabla J(\theta; x^{(i)}, y^{(i)})
   $$

3. **Mini-Batch Gradient Descent**:
   - Uses a mini-batch of training examples to compute the gradient.
   - Update rule:

   $$
   \theta := \theta - \alpha \nabla J(\theta; \{x^{(i)}, y^{(i)}\}_{i=1}^k)
   $$

   Where $k$ is the batch size.

#### &#128161; Visualization
Consider a simple 2D cost function plot where the x-axis represents the model parameter $\theta$ and the y-axis represents the cost $J(\theta)$. Gradient Descent can be visualized as taking steps downhill towards the minimum point on this surface.

### &#128640; <span style="color:#4682b4;">Example 1: Finding the Minimum of a Simple Quadratic Function</span>

In this example, we explore finding the minimum of the function:

$$
y = (x - 3)^2
$$

#### Analytical Solution
The function is a simple quadratic equation, and its minimum can be found analytically. The derivative of the function is:

$$
\frac{dy}{dx} = 2(x - 3)
$$

Setting the derivative to zero to find the critical points:

$$
2(x - 3) = 0 \implies x = 3
$$

Thus, the minimum of the function is at $x = 3$.

#### Gradient Descent Solution
Although we have a closed-form solution, we use Gradient Descent to find the minimum to understand the algorithm's application.

**Gradient Descent Algorithm:**

1. Initialize $x$ with a random value $x_0$.
2. Set the learning rate $\alpha$.
3. Repeat until convergence:
   - Compute the gradient: $\nabla f(x) = 2(x - 3)$
   - Update $x$: $x_{new} = x_{old} - \alpha \nabla f(x_{old})$

This process iteratively adjusts $x$ in the direction of the negative gradient, gradually approaching the minimum of the function.

#### Algorithm Steps:

1. **Initialization**:
   - Start with an initial guess $x_0$.
   - Choose a learning rate $\alpha$.

2. **Iteration**:
   - Compute the gradient: $\nabla f(x) = 2(x - 3)$.
   - Update the value of $x$ using the update rule: 
     $$
     x_{new} = x_{old} - \alpha \cdot 2(x_{old} - 3)
     $$

3. **Convergence**:
   - Repeat the iteration until the change in $x$ is smaller than a predefined threshold, indicating convergence to the minimum.

This algorithm demonstrates the iterative approach of Gradient Descent, providing an intuitive understanding of how the method converges to the minimum value of a function.

````python


# Define the function
def f(x):
    return (x - 3) ** 2

# Generate x values
x = np.linspace(0, 6, 400)
y = f(x)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$y = (x - 3)^2$', color='blue')
plt.axvline(x=3, color='red', linestyle='--', label='Minimum at x=3')
plt.axhline(y=0, color='gray', linestyle='--')
plt.axvline(x=0, color='gray', linestyle='--')
plt.scatter(3, 0, color='red', zorder=5)  # Mark the minimum point
plt.title('Plot of $y = (x - 3)^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Define the function and its derivative
def f(x):
    return (x - 3) ** 2

def f_prime(x):
    return 2 * (x - 3)

# Gradient descent function
def gradient_descent(starting_point, learning_rate, max_iterations=1000):
    x = starting_point
    for i in range(max_iterations):
        gradient = f_prime(x)
        x = x - learning_rate * gradient
    return x

# Starting point and learning rates
starting_point = 5
learning_rates = [0.00001, 0.001, 0.01, 0.1, 1.001]

# Perform gradient descent for each learning rate and report the minimum x found
results = {}
for lr in learning_rates:
    min_x = gradient_descent(starting_point, lr)
    results[lr] = round(min_x, 3)

# Print the results 
print("Learning Rate\t|\tMinimum x Found")
print("------------------------------------")
for lr, min_x in results.items():
    print(f"{lr}\t\t|\t{min_x}")
````

### <span style="color:#800080;">Understanding Batch Gradient Descent (BGD) and Stochastic Gradient Descent (SGD)</span>

#### **Batch Gradient Descent (BGD)**

Batch Gradient Descent is a traditional optimization algorithm used to minimize a cost function $ J(\theta) $ by 
iteratively adjusting the model parameters $ \theta $ in the direction of the negative gradient of the entire training dataset.

- **Definition**: In BGD, the gradient of the cost function with respect to the parameters $ \theta $ is computed over the entire training dataset:
  $$
  \nabla_{\theta} J(\theta) = \frac{1}{m} \sum_{i=1}^{m} \nabla_{\theta} J(\theta; x^{(i)}, y^{(i)})
  $$
  where $ (x^{(i)}, y^{(i)}) $ are training examples, $ m $ is the number of training examples, and
    $ \nabla_{\theta} J(\theta; x^{(i)}, y^{(i)}) $ is the gradient of the cost function for a single training example.

- **Purpose**: BGD aims to find the global minimum or a local minimum of the cost function by making precise steps in the direction of the average gradient of the entire dataset. It guarantees convergence to a minimum under convexity assumptions of $ J(\theta) $.

- **Advantages**:
  - Converges to the minimum of $ J(\theta) $ with a well-chosen learning rate.
  - Suitable for small to moderate-sized datasets where the entire dataset can fit into memory.

- **Disadvantages**:
  - Computationally expensive for large datasets due to the need to process all training examples in each iteration.
  - May get stuck in local minima for non-convex cost functions.

#### **Stochastic Gradient Descent (SGD)**

Stochastic Gradient Descent is a variant of gradient descent where the parameters $ \theta $ 
are updated iteratively based on the gradient of the cost function computed on a single training example or a small subset (mini-batch) of the dataset.

- **Definition**: In SGD, the gradient update rule for a single example or mini-batch is:
  $$
  \theta := \theta - \alpha \nabla_{\theta} J(\theta; x^{(i)}, y^{(i)})
  $$
  where $ \alpha $ is the learning rate and $ \nabla_{\theta} J(\theta; x^{(i)}, y^{(i)}) $ is the gradient of the cost function for a single training example or mini-batch.

- **Purpose**: SGD aims to find the minimum of $ J(\theta) $ by taking frequent 
    steps in the direction of the gradient computed on individual examples or mini-batches. It is well-suited for large datasets and non-convex optimization problems.

- **Advantages**:
  - Computationally efficient, especially for large datasets, as it processes one example or mini-batch at a time.
  - Escapes local minima and helps explore the solution space more effectively in non-convex problems.

- **Disadvantages**:
  - May exhibit noisy convergence due to frequent updates and stochastic nature.
  - Requires careful tuning of the learning rate $ \alpha $ and other hyperparameters.

#### **Why We Use Them**

- **Batch Gradient Descent**:
  - Used when the entire dataset can fit into memory.
  - Guarantees convergence to a minimum under convexity assumptions.
  - Provides a precise direction towards the minimum.

- **Stochastic Gradient Descent**:
  - Ideal for large-scale datasets where BGD is computationally impractical.
  - Helps avoid local minima and explore the solution space more effectively.
  - Allows for online learning where the model is updated continuously as new data becomes available.

In summary, both Batch Gradient Descent and Stochastic Gradient Descent are fundamental optimization techniques used in machine learning to minimize cost functions and train models effectively. The choice between them depends on the dataset size, computational resources, and optimization objectives.

### <span style="color:#800080;">Understanding Convexity Assumptions in Optimization</span>

#### **Definition and Mathematical Explanation**

Convexity in the context of optimization refers to a property of functions and sets that significantly simplifies the task of finding global minima or maxima. 

- **Convex Function**: A function $ f(x) $ defined over a convex set $ S $ is convex if, for all $ x, y \in S $ and for all $ \lambda \in [0, 1] $, it satisfies the inequality:
  $$ f(\lambda x + (1-\lambda) y) \leq \lambda f(x) + (1-\lambda) f(y) $$
  
  Geometrically, this inequality implies that the line segment connecting any two points on the graph of $ f(x) $ lies above or on the graph itself.

- **Implications of Convexity**: 
  - **Unique Global Minima**: Convex functions have the property that any local minimum is also a global minimum. This guarantees that iterative optimization algorithms like Gradient Descent will converge to a single global minimum (if it exists).
  - **Efficient Optimization**: Algorithms for convex optimization are well-developed and efficient, ensuring convergence with reasonable computational effort.
  
- **Convex Set**: A set $ S $ is convex if, for any two points $ x, y \in S $ and for any $ \lambda \in [0, 1] $, the point $ \lambda x + (1-\lambda) y $ also belongs to $ S $.

#### **Applications in Optimization**

Convexity assumptions are crucial in optimization theory and algorithms for several reasons:

- **Guaranteed Convergence**: For convex functions, optimization algorithms such as Gradient Descent are guaranteed to converge to the global minimum. This reliability simplifies the problem of finding optimal solutions in many practical scenarios.

- **Mathematical Rigor**: Convexity provides a rigorous framework for analyzing and solving optimization problems, enabling precise theoretical results and efficient computational techniques.

- **Common Examples**: Many commonly encountered functions in machine learning, such as quadratic functions and linear regression with squared error, are convex, making convex optimization methods widely applicable.

#### **Conclusion**

In summary, convexity assumptions play a pivotal role in optimization by ensuring that optimization algorithms converge to global minima efficiently and reliably. Understanding and leveraging convexity in problem formulation and algorithm design is essential for effective and robust optimization in various fields, including machine learning and operations research.

### <span style="color:#800080;">Understanding Epoch in Gradient Descent</span>

In the context of Gradient Descent, an **epoch** refers to a single pass of the entire dataset through the learning algorithm. During training, the dataset is divided into batches, and each batch is used to compute the gradient and update the model parameters iteratively. 

#### Key Points:

- **Batch Gradient Descent vs. Epoch:**
  - **Batch Gradient Descent:** Computes the gradient using the entire dataset in each iteration. It guarantees a precise estimate of the gradient but can be computationally expensive for large datasets.
  - **Epoch:** Represents one complete cycle of training where the algorithm iterates over the entire dataset once. It consists of multiple iterations where each iteration processes one batch.

- **Significance of Epoch:**
  - The number of epochs is a hyperparameter that defines how many times the learning algorithm will work through the entire training dataset. Each epoch allows the model to learn from the entire dataset, improving parameter estimation with each cycle.
  - After completing an epoch, the model parameters are adjusted based on the accumulated gradients from all batches processed during that epoch.

- **Practical Implementation:**
  - **Training Process:** Typically, training involves running multiple epochs until the model converges or a stopping criterion is met (e.g., achieving a certain level of accuracy or loss reduction).
  - **Mini-Batch Gradient Descent:** Often used in practice, where batches (mini-batches) of data are processed in each iteration within an epoch to balance computational efficiency and gradient accuracy.

Understanding epochs in Gradient Descent helps practitioners manage training cycles effectively, optimizing model performance and convergence towards optimal parameter values.

### &#128640; <span style="color:#4682b4;">Example 4: Estimating Parameters of a Normal Distribution Using Batch Gradient Descent</span>

In this example, we generate random numbers from a normal distribution with unknown parameters $\mu = 3$ and $\sigma^2 = 4$, and attempt to estimate these parameters using Batch Gradient Descent with mini-batches.

#### Problem Setup
We have a dataset $X = \{x_1, x_2, \ldots, x_n\}$ where each $x_i$ is sampled from $N(\mu, \sigma^2)$. Our objective is to find the values of $\mu$ and $\sigma^2$ that maximize the likelihood of observing the data under this distribution.

#### Analytical Solution
The parameters $\mu$ and $\sigma^2$ can be estimated analytically using the Maximum Likelihood Estimation (MLE) approach:
- **Estimation of $\mu$:** $\hat{\mu} = \frac{1}{n} \sum_{i=1}^{n} x_i$
- **Estimation of $\sigma^2$:** $\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \hat{\mu})^2$

#### Negative Log Likelihood Function
To apply gradient descent, we minimize the negative log likelihood function for the normal distribution:
$$
-\text{log}(\mathcal{L}) = \frac{n}{2} \text{log}(2\pi\sigma^2) + \frac{1}{2\sigma^2} \sum_{i=1}^{n} (x_i - \mu)^2
$$

#### Batch Gradient Descent with Mini-Batches
Despite having closed-form solutions for $\mu$ and $\sigma^2$, we apply Batch Gradient Descent with mini-batches as an exercise to understand its application in estimating parameters.

**Batch Gradient Descent Algorithm:**

1. Initialize $\mu$ and $\sigma^2$ with random values.
2. Set the learning rate $\alpha$ and batch size $B$.
3. Repeat until convergence, for multiple epochs:
   - **Epoch:** A complete pass of the entire dataset $X$ through the algorithm.
   - Shuffle the dataset $X$ at the beginning of each epoch to ensure different mini-batches in each iteration.
   - Divide $X$ into mini-batches of size $B$.
   - For each mini-batch $X_{batch}$:
     - Compute the gradients over the mini-batch:
       $$
       \frac{\partial}{\partial \mu} (-\text{log}(\mathcal{L})) = -\frac{1}{\sigma^2} \sum_{i \in X_{batch}} (x_i - \mu)
       $$
       $$
       \frac{\partial}{\partial \sigma} (-\text{log}(\mathcal{L})) = \frac{B}{2\sigma} - \frac{1}{\sigma^3} \sum_{i \in X_{batch}} (x_i - \mu)^2
       $$
     - Update $\mu$ and $\sigma^2$ using the average of gradients across the mini-batch:
       $$
       \mu_{new} = \mu_{old} + \alpha \cdot \frac{1}{\sigma^2} \sum_{i \in X_{batch}} (x_i - \mu)
       $$
       $$
       \sigma_{new} = \sigma_{old} + \alpha \cdot \left( \frac{-B}{2\sigma} + \frac{1}{2\sigma^3} \sum_{i \in X_{batch}} (x_i - \mu)^2 \right)
       $$

This iterative process adjusts $\mu$ and $\sigma^2$ towards values that maximize the likelihood of observing the data under a normal distribution, using gradients computed over mini-batches. By leveraging mini-batches, Batch Gradient Descent improves computational efficiency while approximating the true parameter values, demonstrating its utility in large-scale optimization tasks.

</div>

### <span style="color:#800080;">The Importance of Stopping Strategies in Gradient Descent</span>

When training models using gradient descent, it is crucial to implement effective stopping strategies. Without these strategies, the optimization process might run indefinitely or lead to issues such as overfitting or divergence. Here are some reasons why we need stopping strategies:

- **Prevent Overfitting**: Continuing to train the model for too long can cause it to overfit to the training data, leading to poor generalization on unseen data.
- **Avoid Divergence**: If the learning rate is not set appropriately, the loss may start to increase after a certain number of iterations, indicating that the model parameters are diverging.
- **Save Computational Resources**: Stopping the training process at the right time can save computational resources and time, making the training process more efficient.
- **Ensure Convergence**: Proper stopping criteria can help ensure that the model parameters converge to a stable solution.

#### Common Stopping Strategies

Here are some common strategies used to determine when to stop the gradient descent process:

1. **Maximum Number of Iterations (Epochs)**:
   - Stop the training process after a predetermined number of iterations or epochs.

2. **Tolerance on Change in Objective Function**:
   - Stop if the change in the objective function (e.g., loss) between consecutive iterations falls below a predefined threshold.

3. **Tolerance on Change in Parameters**:
   - Stop if the change in the model parameters (weights) between consecutive iterations is smaller than a predefined threshold.

4. **Gradient Magnitude**:
   - Stop if the magnitude of the gradient vector falls below a certain level, indicating that further optimization is unlikely to make significant progress.

5. **Validation Error (Early Stopping)**:
   - Monitor the error on a separate validation set and stop if the error starts to increase, which helps prevent overfitting.

By incorporating these stopping strategies, we can make the gradient descent process more robust, efficient, and effective in training models.

</div>


````python
# Generate data from N(3, 4)
np.random.seed(42)  # For reproducibility
mu_true = 3
sigma_true = 2  # Standard deviation is square root of variance
sample_size = 10000
data = np.random.normal(mu_true, sigma_true, size=sample_size)



# Batch Gradient Descent to estimate mu and sigma
def batch_gradient_descent(data, starting_point, learning_rate, batch_size, num_epochs, tolerance=1e-5):
    mu, sigma = starting_point
    n = len(data)
    losses = []
    prev_loss = np.inf
    
    for epoch in range(num_epochs):
        np.random.shuffle(data)  # Shuffle data at the start of each epoch
        total_loss = 0
        
        for i in range(0, n, batch_size):
            batch = data[i:i+batch_size]
            batch_size_current = len(batch)
            
            # Compute gradients
            grad_mu = -np.sum(batch - mu) / (sigma**2)
            grad_sigma = batch_size_current /  sigma - np.sum((batch - mu)**2) / (sigma**3)
            
            # Update parameters
            mu -= learning_rate * grad_mu
            sigma -= learning_rate * grad_sigma
            
            # Compute loss (negative log likelihood)
            loss = np.sum((np.log(2 * np.pi * sigma**2) / 2) + ((batch - mu)**2 / (2 * sigma**2)))
            total_loss += loss
        
        losses.append(total_loss)
        
        # Stopping criteria based on loss improvement
        if abs(prev_loss - total_loss) < tolerance:
            print(f"Convergence achieved at epoch {epoch}")
            break
        
        prev_loss = total_loss
    
    return mu, sigma, losses

# Initialize parameters
starting_point = (1, 1)
learning_rate = 0.0001
batch_size = 5
num_epochs = 20

# Perform batch gradient descent
estimated_mu, estimated_sigma, losses = batch_gradient_descent(data, starting_point, learning_rate, batch_size, num_epochs)

# Print the estimated values
print(f"Estimated mu: {estimated_mu:.4f}")
print(f"Estimated sigma: {estimated_sigma:.4f}")

# Plot the loss curve
plt.figure(figsize=(8, 4.5))
plt.plot(losses, label='Loss', marker = 'o' , color = 'purple' , linewidth = 2 , markersize = 8 ,ls='--')
plt.title('Batch Gradient Descent: Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Negative Log Likelihood')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

### <span style="color:#800080;">Backward Propagation in Neural Networks</span>

Backward propagation, or backpropagation, is a critical process in training neural networks. It involves computing the gradient of the loss function with respect to each weight by the chain rule, iteratively applying it from the output layer back to the input layer. This process allows us to update the model parameters to minimize the loss function.

### Steps of Backpropagation:

1. **Compute the Loss:**
   - After a forward pass, compute the loss $ L $ using the predicted output $ \hat{Y} $ and the true labels $ Y $.
   $$
   L = -\frac{1}{N} \sum_{i=1}^{N} \sum_{j=1}^{3} y_{ij} \log(\hat{y}_{ij})
   $$

2. **Calculate the Gradient of the Loss with Respect to Output:**
   - For the output layer, compute the gradient of the loss with respect to the logits $ Z^{(2)} $.
   $$
   \frac{\partial L}{\partial Z^{(2)}} = \hat{Y} - Y
   $$

3. **Propagate the Gradient Back to the Hidden Layer:**
   - Compute the gradient of the loss with respect to the weights and biases of the output layer.
   $$
   \frac{\partial L}{\partial W^{(2)}} = H^{(1)^T} \cdot \frac{\partial L}{\partial Z^{(2)}}
   $$
   $$
   \frac{\partial L}{\partial b^{(2)}} = \sum_{i=1}^{N} \frac{\partial L}{\partial Z^{(2)}_i}
   $$

   - Propagate the gradient back to the hidden layer:
   $$
   \frac{\partial L}{\partial H^{(1)}} = \frac{\partial L}{\partial Z^{(2)}} \cdot W^{(2)^T}
   $$

4. **Apply ReLU Derivative:**
   - Compute the gradient through the ReLU activation function:
   $$
   \frac{\partial L}{\partial A^{(1)}} = \frac{\partial L}{\partial H^{(1)}} \odot \text{ReLU}'(A^{(1)})
   $$
   where $ \odot $ denotes element-wise multiplication and $ \text{ReLU}'(x) = \begin{cases} 
   1 & \text{if } x > 0 \\
   0 & \text{if } x \leq 0 
   \end{cases} $

5. **Propagate the Gradient Back to the Input Layer:**
   - Compute the gradient of the loss with respect to the weights and biases of the hidden layer.
   $$
   \frac{\partial L}{\partial W^{(1)}} = X^T \cdot \frac{\partial L}{\partial A^{(1)}}
   $$
   $$
   \frac{\partial L}{\partial b^{(1)}} = \sum_{i=1}^{N} \frac{\partial L}{\partial A^{(1)}_i}
   $$

6. **Update the Weights and Biases:**
   - Using the gradients computed, update the weights and biases using gradient descent:
   $$
   W^{(1)} \leftarrow W^{(1)} - \alpha \cdot \frac{\partial L}{\partial W^{(1)}}
   $$
   $$
   b^{(1)} \leftarrow b^{(1)} - \alpha \cdot \frac{\partial L}{\partial b^{(1)}}
   $$
   $$
   W^{(2)} \leftarrow W^{(2)} - \alpha \cdot \frac{\partial L}{\partial W^{(2)}}
   $$
   $$
   b^{(2)} \leftarrow b^{(2)} - \alpha \cdot \frac{\partial L}{\partial b^{(2)}}
   $$
   where $ \alpha $ is the learning rate.

### Summary:

- **Forward Pass:** Calculate the predictions and loss.
- **Backward Pass:** Compute the gradients of the loss with respect to each parameter.
- **Parameter Update:** Adjust the parameters in the direction that minimizes the loss.

By iteratively performing these steps, the neural network learns to adjust its weights and biases to minimize the loss function, thereby improving its performance on the training data.

</div>


````python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# One-hot encode the target labels
encoder = OneHotEncoder(sparse=False)
y_onehot = encoder.fit_transform(y.reshape(-1, 1))

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_onehot, test_size=0.2, random_state=42)

# Define activation functions and their derivatives
def relu(Z):
    return np.maximum(0, Z)

def relu_derivative(Z):
    return Z > 0

def softmax(Z):
    expZ = np.exp(Z - np.max(Z, axis=1, keepdims=True))
    return expZ / expZ.sum(axis=1, keepdims=True)

# Define the neural network architecture
input_size = 4
hidden_size = 5
output_size = 3

# Initialize weights and biases
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

# Training hyperparameters
learning_rate = 0.01
num_epochs = 10

# Training loop
losses = []
for epoch in range(num_epochs):
    # Forward pass
    Z1 = X_train.dot(W1) + b1
    A1 = relu(Z1)
    Z2 = A1.dot(W2) + b2
    A2 = softmax(Z2)
    
    # Compute the loss (cross-entropy)
    loss = -np.mean(np.sum(y_train * np.log(A2), axis=1))
    losses.append(loss)
    
    # Backward pass
    dZ2 = A2 - y_train
    dW2 = A1.T.dot(dZ2)
    db2 = np.sum(dZ2, axis=0, keepdims=True)
    dA1 = dZ2.dot(W2.T)
    dZ1 = dA1 * relu_derivative(Z1)
    dW1 = X_train.T.dot(dZ1)
    db1 = np.sum(dZ1, axis=0, keepdims=True)
    
    # Update weights and biases
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

# Plot the loss curve
plt.plot(losses , marker = 'o' , color ='darkblue')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.grid(True)
plt.show()

# Evaluate the model
def predict(X):
    Z1 = X.dot(W1) + b1
    A1 = relu(Z1)
    Z2 = A1.dot(W2) + b2
    A2 = softmax(Z2)
    return np.argmax(A2, axis=1)

y_pred = predict(X_test)
y_test_labels = np.argmax(y_test, axis=1)

# Classification report
print("Classification Report:")
print(classification_report(y_test_labels, y_pred, target_names=iris.target_names))

# Confusion matrix
conf_matrix = confusion_matrix(y_test_labels, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Visualize the confusion matrix
plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(iris.target_names))
plt.xticks(tick_marks, iris.target_names, rotation=45)
plt.yticks(tick_marks, iris.target_names)

thresh = conf_matrix.max() / 2
for i, j in np.ndindex(conf_matrix.shape):
    plt.text(j, i, format(conf_matrix[i, j], 'd'),
             ha='center', va='center',
             color='white' if conf_matrix[i, j] > thresh else 'black')

plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.tight_layout()
plt.show()
````

### &#128640; <span style="color:#4682b4;">AdaGrad Optimization Algorithm</span>

AdaGrad (short for Adaptive Gradient Algorithm) is an optimization method that adapts the learning rate for each parameter individually during training. It is particularly effective for dealing with sparse data and for ensuring that parameters with infrequent updates are not overwhelmed by the updates of frequently updated parameters.

#### Key Concepts

1. **Per-Parameter Learning Rate**:
   - Adjusts the learning rate for each parameter based on the historical sum of squared gradients for that parameter.

2. **Learning Rate Adaptation**:
   - Parameters with large gradients receive smaller updates, while parameters with small gradients receive larger updates.

#### Algorithm Steps

1. **Initialization**:
   - Initialize parameters $\theta_0$ (weights of the neural network).
   - Initialize a sum of squared gradients accumulator $G_0 = 0$.
   - Set the global learning rate $\alpha$ and a small constant $\epsilon$ to avoid division by zero.

2. **Update Rule**:
   - For each iteration $t$:
     - Compute the gradient $g_t = \nabla_\theta f_t(\theta_{t-1})$.
     - Update the sum of squared gradients:
       $$
       G_t = G_{t-1} + g_t^2
       $$
     - Adjust the parameters using the adapted learning rate:
       $$
       \theta_t = \theta_{t-1} - \frac{\alpha}{\sqrt{G_t + \epsilon}} \cdot g_t
       $$

#### Advantages of AdaGrad

- **Adaptive Learning Rates**: By adjusting the learning rate based on the history of gradients, AdaGrad allows for more refined updates.
- **Effective for Sparse Data**: AdaGrad is particularly useful for problems with sparse features, as it can perform larger updates for infrequent parameters.
- **No Need for Learning Rate Decay**: The learning rate automatically decreases over time without the need for manually tuning a decay schedule.

#### Common Hyperparameters

- $\alpha$ (initial learning rate): Typical values are $0.01$, $0.1$, etc.
- $\epsilon$ (smoothing term): A common value is $10^{-8}$.

### &#128640; <span style="color:#4682b4;">RMSProp Optimization Algorithm</span>

RMSProp (Root Mean Square Propagation) is an optimization method designed to address the problem of AdaGrad's rapidly decreasing learning rate. By maintaining a moving average of the squared gradients, RMSProp ensures that the learning rate does not decrease too quickly, leading to more stable and efficient training.

#### Key Concepts

1. **Exponential Decay Average**:
   - Maintains an exponentially decaying average of past squared gradients to adapt the learning rate for each parameter.

2. **Decay Rate**:
   - Uses a decay rate to control the influence of past gradients, ensuring that the learning rate remains balanced over time.

#### Algorithm Steps

1. **Initialization**:
   - Initialize parameters $\theta_0$ (weights of the neural network).
   - Initialize a moving average of squared gradients $E[g^2]_0 = 0$.
   - Set the global learning rate $\alpha$, decay rate $\rho$, and a small constant $\epsilon$ to avoid division by zero.

2. **Update Rule**:
   - For each iteration $t$:
     - Compute the gradient $g_t = \nabla_\theta f_t(\theta_{t-1})$.
     - Update the moving average of squared gradients:
       $$
       E[g^2]_t = \rho \cdot E[g^2]_{t-1} + (1 - \rho) \cdot g_t^2
       $$
     - Adjust the parameters using the adapted learning rate:
       $$
       \theta_t = \theta_{t-1} - \frac{\alpha}{\sqrt{E[g^2]_t + \epsilon}} \cdot g_t
       $$

#### Advantages of RMSProp

- **Stabilized Learning Rate**: By using an exponentially decaying average, RMSProp prevents the learning rate from decreasing too rapidly.
- **Effective for Non-Stationary Objectives**: Suitable for tasks where the loss landscape changes over time.
- **Balances Learning Rate Adaptation**: Maintains a balance between AdaGradâ€™s fast adaptation and the need for a consistent learning rate.

#### Common Hyperparameters

- $\alpha$ (initial learning rate): Typical values are $0.001$, $0.01$, etc.
- $\rho$ (decay rate): Common values range from $0.9$ to $0.99$.
- $\epsilon$ (smoothing term): A common value is $10^{-8}$.

### &#128640; <span style="color:#4682b4;">Adam Optimization Algorithm</span>

Adam (short for Adaptive Moment Estimation) is an advanced optimization algorithm used to train machine learning models, particularly deep neural networks. Adam combines the benefits of two other extensions of stochastic gradient descent: AdaGrad and RMSProp.

#### Key Concepts

1. **Exponential Moving Average (EMA)**:
   - EMA of the first moment (mean) of the gradients.
   - EMA of the second moment (uncentered variance) of the gradients.

2. **Bias Correction**:
   - Corrects the biases in the moving averages, especially during the initial steps of training.

#### Algorithm Steps

1. **Initialization**:
   - Initialize parameters $\theta_0$ (weights of the neural network).
   - Initialize the first moment vector $m_0 = 0$.
   - Initialize the second moment vector $v_0 = 0$.
   - Set hyperparameters $\alpha$ (learning rate), $\beta_1$ (decay rate for the first moment estimate), $\beta_2$ (decay rate for the second moment estimate), and $\epsilon$ (a small constant to avoid division by zero).

2. **Update Rule**:
   - For each iteration $t$:
     - Compute the gradients $g_t = \nabla_\theta f_t(\theta_{t-1})$.
     - Update biased first moment estimate:
       $$
       m_t = \beta_1 \cdot m_{t-1} + (1 - \beta_1) \cdot g_t
       $$
     - Update biased second moment estimate:
       $$
       v_t = \beta_2 \cdot v_{t-1} + (1 - \beta_2) \cdot g_t^2
       $$
     - Compute bias-corrected first moment estimate:
       $$
       \hat{m}_t = \frac{m_t}{1 - \beta_1^t}
       $$
     - Compute bias-corrected second moment estimate:
       $$
       \hat{v}_t = \frac{v_t}{1 - \beta_2^t}
       $$
     - Update parameters:
       $$
       \theta_t = \theta_{t-1} - \alpha \cdot \frac{\hat{m}_t}{\sqrt{\hat{v}_t} + \epsilon}
       $$

#### Advantages of Adam

- **Adaptive Learning Rates**: Adam adjusts the learning rates for each parameter individually, which can lead to faster convergence.
- **Combines Advantages of AdaGrad and RMSProp**: Adam combines the per-parameter learning rate adjustment of AdaGrad with the momentum-like updates of RMSProp.
- **Bias Correction**: Adam includes bias correction to account for the initialization of the moment estimates, which is particularly important in the early stages of training.

#### Common Hyperparameters

- $\alpha$ (learning rate): Typical values are $0.001$, $0.0001$, etc.
- $\beta_1$ (first moment decay rate): A common value is $0.9$.
- $\beta_2$ (second moment decay rate): A common value is $0.999$.
- $\epsilon$ (smoothing term): A common value is $10^{-8}$.

#### Example Usage in Python (with TensorFlow/Keras)

```python
